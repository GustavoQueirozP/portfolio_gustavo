```python
# Passo 1: Vit√≥rias por piloto por ano
results_drivers = results.merge(drivers, left_on='driverId', right_on='driverId')
results_drivers_races = results_drivers.merge(races[['raceId', 'year']], on='raceId')

vit√≥rias = results_drivers_races[results_drivers_races['positionOrder'] == 1]

wins_piloto_ano = (
    vit√≥rias.groupby(['year', 'driverId', 'surname', 'constructorId'])
    .size()
    .reset_index(name='vit√≥rias')
)

# Passo 2: Vit√≥rias por equipe por ano
results_constructors = results.merge(constructors, left_on='constructorId', right_on='constructorId')
results_constructors_races = results_constructors.merge(races[['raceId', 'year']], on='raceId')

vit√≥rias_equipe = results_constructors_races[results_constructors_races['positionOrder'] == 1]

wins_equipe_ano = (
    vit√≥rias_equipe.groupby(['year', 'constructorId', 'name'])
    .size()
    .reset_index(name='vit√≥rias')
)

# Passo 3: Melhor piloto por ano (maior n√∫mero de vit√≥rias)
melhor_piloto_ano = (
    wins_piloto_ano.loc[
        wins_piloto_ano.groupby('year')['vit√≥rias'].idxmax()
    ]
    .reset_index(drop=True)
)

# Passo 4: Melhor equipe por ano (maior n√∫mero de vit√≥rias)
melhor_equipe_ano = (
    wins_equipe_ano.loc[
        wins_equipe_ano.groupby('year')['vit√≥rias'].idxmax()
    ]
    .reset_index(drop=True)
)

# Passo 5: Comparar se o melhor piloto estava na melhor equipe no mesmo ano
comparacao = melhor_piloto_ano.merge(
    melhor_equipe_ano,
    left_on='year',
    right_on='year',
    suffixes=('_piloto', '_equipe')
)

# Criar coluna booleana para verificar se o piloto correu na melhor equipe
comparacao['piloto_na_melhor_equipe'] = comparacao['constructorId_piloto'] == comparacao['constructorId_equipe']

# Mostrar anos em que o melhor piloto N√ÉO estava na melhor equipe
anos_diferentes = comparacao[comparacao['piloto_na_melhor_equipe'] == False]

print(anos_diferentes[['year', 'surname', 'name', 'vit√≥rias_piloto', 'vit√≥rias_equipe', 'piloto_na_melhor_equipe']])
```

```python
# Filtrar corridas finalizadas
corridas_validas = results[(results['positionOrder'].notnull()) & (results['positionOrder'] > 0)]

# Calcular posi√ß√µes ganhas (grid - posi√ß√£o final)
corridas_validas['posicoes_ganhas'] = corridas_validas['grid'] - corridas_validas['positionOrder']

# Ignorar largadas no grid 0 (n√£o informadas)
corridas_validas = corridas_validas[corridas_validas['grid'] > 0]

# M√©dia de posi√ß√µes ganhas por piloto
media_posicoes_ganhas = corridas_validas.groupby('driverId')['posicoes_ganhas'].mean().reset_index()

# Juntar com nomes dos pilotos
media_posicoes_ganhas = media_posicoes_ganhas.merge(drivers[['driverId', 'surname']], on='driverId')

# Top 10 pilotos que mais ganharam posi√ß√µes em m√©dia
top10_ganhos = media_posicoes_ganhas.sort_values(by='posicoes_ganhas', ascending=False).head(10)

# Gr√°fico
plt.figure(figsize=(12,6))
sns.barplot(x='posicoes_ganhas', y='surname', data=top10_ganhos, color='steelblue')
plt.title('Top 10 Pilotos com Maior M√©dia de Posi√ß√µes Ganhas por Corrida')
plt.xlabel('Posi√ß√µes Ganhas em M√©dia')
plt.ylabel('Piloto')
plt.tight_layout()
plt.show()
```

```python
# Filtrar apenas os vencedores de cada corrida
vencedores = results[results['positionOrder'] == 1]

# Juntar com drivers para pegar nome do piloto
vencedores = vencedores.merge(drivers[['driverId', 'surname', 'forename']], on='driverId')

# Juntar com races para pegar dados da corrida (ano, nome)
vencedores = vencedores.merge(races[['raceId', 'year', 'name']], on='raceId')

# Ordenar pela posi√ß√£o de largada (grid), da mais atrasada para a mais a frente
vencedores_mais_atras = vencedores.sort_values(by='grid', ascending=False)

# Selecionar as colunas mais relevantes para a tabela final
tabela_pilotos_atras_ganhadores = vencedores_mais_atras[['year', 'name', 'forename', 'surname', 'grid']]

# Mostrar os 10 primeiros resultados
print(tabela_pilotos_atras_ganhadores.head(10))
```

```python
# 1. Vit√≥rias por piloto
vit√≥rias = results[results['positionOrder'] == 1].groupby('driverId').size().rename('vit√≥rias')

# 2. Posi√ß√µes ganhas: grid - positionOrder (apenas corridas que terminou com posi√ß√£o v√°lida)
results['posicoes_ganhas'] = results['grid'] - results['positionOrder']
# Somar posi√ß√µes ganhas por piloto
posicoes_ganhas = results.groupby('driverId')['posicoes_ganhas'].sum()

# 3. Total de corridas disputadas
total_corridas = results.groupby('driverId').size().rename('total_corridas')

# 4. Qualidade da equipe (m√©dia de vit√≥rias da equipe por ano nos anos que piloto correu)
# Primeiro, pegar vit√≥rias por equipe por ano
vit√≥rias_equipe_ano = results[results['positionOrder'] == 1].groupby(['constructorId', 'raceId']).size().reset_index(name='vit√≥rias')
vit√≥rias_equipe_ano = vit√≥rias_equipe_ano.merge(races[['raceId', 'year']], on='raceId')
vit√≥rias_equipe_ano = vit√≥rias_equipe_ano.groupby(['constructorId', 'year'])['vit√≥rias'].sum().reset_index()

# Pegar os anos e equipes de cada piloto
piloto_equipe_anos = results[['driverId', 'constructorId', 'raceId']].merge(races[['raceId', 'year']], on='raceId')
piloto_equipe_anos = piloto_equipe_anos.drop_duplicates()

# Juntar com vit√≥rias da equipe por ano
piloto_equipe_anos = piloto_equipe_anos.merge(vit√≥rias_equipe_ano, on=['constructorId', 'year'], how='left')

# Calcular m√©dia de vit√≥rias da equipe do piloto (indicador de for√ßa da equipe)
qualidade_equipe = piloto_equipe_anos.groupby('driverId')['vit√≥rias'].mean().fillna(0).rename('qualidade_equipe')

# 5. Combinar tudo em um dataframe
df_ranking = pd.concat([vit√≥rias, posicoes_ganhas, total_corridas, qualidade_equipe], axis=1).fillna(0)

# Normalizar para balancear as m√©tricas (opcional, usando min-max)
df_norm = (df_ranking - df_ranking.min()) / (df_ranking.max() - df_ranking.min())

# Definir pesos para cada fator (pode ajustar conforme preferir)
pesos = {
    'vit√≥rias': 0.5,
    'posicoes_ganhas': 0.3,
    'total_corridas': 0.1,
    'qualidade_equipe': 0.1
}

# Calcular score final ponderado
df_norm['score_final'] = (
    df_norm['vit√≥rias'] * pesos['vit√≥rias'] +
    df_norm['posicoes_ganhas'] * pesos['posicoes_ganhas'] +
    df_norm['total_corridas'] * pesos['total_corridas'] +
    df_norm['qualidade_equipe'] * pesos['qualidade_equipe']
)

# Ordenar pelo score e juntar com nomes dos pilotos
df_ranking_final = df_norm.sort_values(by='score_final', ascending=False)
df_ranking_final = df_ranking_final.merge(drivers[['driverId', 'forename', 'surname']], on='driverId')

# Mostrar top 10
top10_pilotos = df_ranking_final[['forename', 'surname', 'score_final']].head(10)
print(top10_pilotos)
```

```python
# 1. Vit√≥rias por piloto
vit√≥rias = results[results['positionOrder'] == 1].groupby('driverId').size().rename('vit√≥rias')

# 2. Pontos conquistados por piloto
pontos = results.groupby('driverId')['points'].sum()

# 3. Quantidade de corridas disputadas
total_corridas = results.groupby('driverId').size().rename('total_corridas')

# 4. Abandonos (DNF) por piloto (statusId diferente de 1 indica abandono ou outro status)
abandono = results[results['statusId'] != 1].groupby('driverId').size().rename('abandono')

# 5. Qualidade da equipe: m√©dia de vit√≥rias da equipe no ano do piloto (mesma l√≥gica da resposta anterior)
vit√≥rias_equipe_ano = results[results['positionOrder'] == 1].groupby(['constructorId', 'raceId']).size().reset_index(name='vit√≥rias')
vit√≥rias_equipe_ano = vit√≥rias_equipe_ano.merge(races[['raceId', 'year']], on='raceId')
vit√≥rias_equipe_ano = vit√≥rias_equipe_ano.groupby(['constructorId', 'year'])['vit√≥rias'].sum().reset_index()

piloto_equipe_anos = results[['driverId', 'constructorId', 'raceId']].merge(races[['raceId', 'year']], on='raceId').drop_duplicates()
piloto_equipe_anos = piloto_equipe_anos.merge(vit√≥rias_equipe_ano, on=['constructorId', 'year'], how='left')
qualidade_equipe = piloto_equipe_anos.groupby('driverId')['vit√≥rias'].mean().fillna(0).rename('qualidade_equipe')

# 6. M√©dia da posi√ß√£o de largada (grid)
media_largada = results.groupby('driverId')['grid'].mean().rename('media_largada')

# 7. Total de posi√ß√µes ganhas (grid - positionOrder)
results['posicoes_ganhas'] = results['grid'] - results['positionOrder']
posicoes_ganhas = results.groupby('driverId')['posicoes_ganhas'].sum().rename('posicoes_ganhas')

# Unir tudo em um √∫nico DataFrame
df = pd.concat([vit√≥rias, pontos, total_corridas, abandono, qualidade_equipe, media_largada, posicoes_ganhas], axis=1).fillna(0)

# Normalizar as colunas (min-max) para balancear escala
df_norm = (df - df.min()) / (df.max() - df.min())

# Ajustar o sinal da coluna abandono (penalizar)
df_norm['abandono'] = -df_norm['abandono']

# Ajustar o sinal da posi√ß√£o de largada (quanto menor, melhor)
df_norm['media_largada'] = 1 - df_norm['media_largada']

# Definir pesos
pesos = {
    'vit√≥rias': 0.35,
    'points': 0.25,
    'posicoes_ganhas': 0.15,
    'qualidade_equipe': 0.10,
    'total_corridas': 0.05,
    'media_largada': 0.05,
    'abandono': 0.05
}

# Calcular score final
df_norm['score_final'] = (
    df_norm['vit√≥rias'] * pesos['vit√≥rias'] +
    df_norm['points'] * pesos['points'] +
    df_norm['posicoes_ganhas'] * pesos['posicoes_ganhas'] +
    df_norm['qualidade_equipe'] * pesos['qualidade_equipe'] +
    df_norm['total_corridas'] * pesos['total_corridas'] +
    df_norm['media_largada'] * pesos['media_largada'] +
    df_norm['abandono'] * pesos['abandono']
)

# Juntar com os nomes dos pilotos
df_final = df_norm.sort_values(by='score_final', ascending=False)
df_final = df_final.merge(drivers[['driverId', 'forename', 'surname']], on='driverId')

# Mostrar top 10
top10 = df_final[['forename', 'surname', 'score_final']].head(10)
print(top10)
```

```python
import numpy as np

# Exemplo: adicionar coluna year (se n√£o existir, junte com races para pegar ano)
results = results.merge(races[['raceId','year']], on='raceId', how='left')

# Abandono: statusId diferente de 1 (1 = terminou a corrida)
results['abandonou'] = results['statusId'] != 1

# Posi√ß√µes ganhas na corrida
results['posicoes_ganhas'] = results['grid'] - results['positionOrder']

# Corridas pontuadas (considerando pontos > 0)
results['pontuou'] = results['points'] > 0

# Agrupamento por piloto e ano para calcular m√©tricas
metrics = results.groupby(['driverId', 'year']).agg(
    corridas = ('raceId', 'count'),
    vit√≥rias = ('positionOrder', lambda x: (x == 1).sum()),
    poles = ('grid', lambda x: (x == 1).sum()),  # pole = grid 1
    abandonos = ('abandonou', 'sum'),
    pontos = ('points', 'sum'),
    pos_media_largada = ('grid', 'mean'),
    pos_media_chegada = ('positionOrder', 'mean'),
    posicoes_ganhas_media = ('posicoes_ganhas', 'mean'),
    pct_corridas_top10 = ('positionOrder', lambda x: (x <= 10).mean())
).reset_index()
```

```python
metrics['pontos_por_corrida'] = metrics['pontos'] / metrics['corridas']
```

```python
# Calcula pontos por equipe e ano
equipe_metrics = results.groupby(['constructorId', 'year']).agg(
    pontos_equipe = ('points', 'sum'),
    corridas_equipe = ('raceId', 'count')
).reset_index()

equipe_metrics['pontos_equipe_por_corrida'] = equipe_metrics['pontos_equipe'] / equipe_metrics['corridas_equipe']

# Juntar info equipe no metrics piloto
metrics = metrics.merge(
    results[['driverId','constructorId','year']].drop_duplicates(),
    on=['driverId','year'], how='left'
).merge(
    equipe_metrics[['constructorId','year','pontos_equipe_por_corrida']],
    on=['constructorId','year'],
    how='left'
)
```

```python
# Pesos arbitr√°rios (ajuste como quiser)
w_corridas = 0.15
w_vitorias = 0.3
w_pontos = 0.25
w_abandonos = -0.2
w_poles = 0.05
w_posicoes_ganhas = 0.1
w_equipe = 0.2
w_top10 = 0.1

# Escalar m√©tricas para 0-1 (min-max)
def min_max_scaling(series):
    return (series - series.min()) / (series.max() - series.min())

metrics['vit√≥rias_scaled'] = min_max_scaling(metrics['vit√≥rias'])
metrics['pontos_scaled'] = min_max_scaling(metrics['pontos_por_corrida'])
metrics['abandonos_scaled'] = min_max_scaling(metrics['abandonos'])
metrics['poles_scaled'] = min_max_scaling(metrics['poles'])
metrics['posicoes_ganhas_scaled'] = min_max_scaling(metrics['posicoes_ganhas_media'])
metrics['equipe_scaled'] = min_max_scaling(metrics['pontos_equipe_por_corrida'])
metrics['top10_scaled'] = min_max_scaling(metrics['pct_corridas_top10'])
metrics['corridas_scaled'] = min_max_scaling(metrics['corridas'])

# √çndice de desempenho
metrics['indice_desempenho'] = (
    w_corridas * metrics['corridas_scaled'] +
    w_vitorias * metrics['vit√≥rias_scaled'] +
    w_pontos * metrics['pontos_scaled'] +
    w_abandonos * (1 - metrics['abandonos_scaled']) +  # menos abandonos = melhor
    w_poles * metrics['poles_scaled'] +
    w_posicoes_ganhas * metrics['posicoes_ganhas_scaled'] +
    w_equipe * metrics['equipe_scaled'] +
    w_top10 * metrics['top10_scaled']
)
```

```python
ranking_pilotos = metrics.groupby('driverId').agg(
    indice_medio = ('indice_desempenho', 'mean'),
    total_corridas = ('corridas', 'sum'),
    total_vitorias = ('vit√≥rias', 'sum'),
    total_pontos = ('pontos', 'sum')
).reset_index()

# Juntar nomes
ranking_pilotos = ranking_pilotos.merge(drivers[['driverId', 'surname']], on='driverId', how='left')

# Top 10
top10 = ranking_pilotos.sort_values('indice_medio', ascending=False).head(10)
print(top10[['surname','indice_medio','total_corridas','total_vitorias','total_pontos']])
```

```python
import matplotlib.pyplot as plt

# Selecionar as colunas de interesse para o gr√°fico
cols_componentes = [
    'corridas_scaled', 'vit√≥rias_scaled', 'pontos_scaled',
    'abandonos_scaled', 'poles_scaled', 'posicoes_ganhas_scaled',
    'equipe_scaled', 'top10_scaled'
]

# Pesos aplicados para cada componente (positivo ou negativo)
pesos = {
    'corridas_scaled': w_corridas,
    'vit√≥rias_scaled': w_vitorias,
    'pontos_scaled': w_pontos,
    'abandonos_scaled': -w_abandonos,  # lembre que para abandonos usamos 1 - scaled, ent√£o aqui s√≥ plota normal
    'poles_scaled': w_poles,
    'posicoes_ganhas_scaled': w_posicoes_ganhas,
    'equipe_scaled': w_equipe,
    'top10_scaled': w_top10
}

# Criar DataFrame apenas com o top10 pilotos
top10_componentes = metrics[metrics['driverId'].isin(top10['driverId'])]

# Aplicar os pesos aos componentes
for col in cols_componentes:
    if col == 'abandonos_scaled':
        # para abandonar menos, usamos 1 - scaled
        top10_componentes[col] = 1 - top10_componentes[col]
    top10_componentes[col] = top10_componentes[col] * pesos[col]

# Agrupar por piloto somando as contribui√ß√µes (m√©dia por piloto)
top10_comp_agg = top10_componentes.groupby('driverId')[cols_componentes].mean()

# Juntar nomes dos pilotos para o eixo y
top10_comp_agg = top10_comp_agg.merge(drivers[['driverId', 'surname']], on='driverId')
top10_comp_agg = top10_comp_agg.set_index('surname')

# Ordenar pelo √≠ndice de desempenho m√©dio para ficar igual o ranking
top10_comp_agg = top10_comp_agg.loc[top10['surname']]

# Plotando o gr√°fico de barras empilhadas
fig, ax = plt.subplots(figsize=(12, 7))

bottoms = np.zeros(len(top10_comp_agg))
labels = {
    'corridas_scaled': 'Quantidade de Corridas',
    'vit√≥rias_scaled': 'Vit√≥rias',
    'pontos_scaled': 'Pontos por Corrida',
    'abandonos_scaled': 'Menos Abandonos',
    'poles_scaled': 'Poles',
    'posicoes_ganhas_scaled': 'Posi√ß√µes Ganhas',
    'equipe_scaled': 'Qualidade da Equipe',
    'top10_scaled': '% Corridas no Top 10'
}

for col in cols_componentes:
    ax.barh(
        top10_comp_agg.index,
        top10_comp_agg[col],
        left=bottoms,
        label=labels[col]
    )
    bottoms += top10_comp_agg[col]

ax.set_xlabel('Contribui√ß√£o para o √çndice de Desempenho')
ax.set_title('Componentes do √çndice de Desempenho dos 10 Melhores Pilotos')
ax.legend(loc='lower right')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
