```python
import pandas as pd
import os
```

```python
# Caminho da pasta onde est√£o os arquivos CSV
caminho = r"C:\Users\gusta\portfolio_gustavo\projetos\analise_ifood"

# Lista de arquivos
arquivos = ['channels.csv', 'deliveries.csv', 'drivers.csv', 'hubs.csv', 
            'orders.csv', 'payments.csv', 'stores.csv']
dfs = {}

# Carregando cada arquivo
for arquivo in arquivos:
    nome = arquivo.replace('.csv', '')
    try:
        dfs[nome] = pd.read_csv(os.path.join(caminho, arquivo), encoding='latin1')
        print(f"{nome} carregado com sucesso. Formato: {dfs[nome].shape}")
    except Exception as e:
        print(f"Erro ao carregar {arquivo}: {e}")
```

```python
for nome, df in dfs.items():
    print(f"\n===== {nome.upper()} =====")
    print("Colunas:", df.columns.tolist())
    print("Formato:", df.shape)
    display(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Tamanho padr√£o dos gr√°ficos
plt.rcParams["figure.figsize"] = (8, 5)

# Contagem de pedidos por status
status_counts = dfs["orders"]["order_status"].value_counts().reset_index()
status_counts.columns = ['order_status', 'count']

# Gr√°fico
sns.barplot(data=status_counts, x='order_status', y='count', palette="viridis", hue='order_status', legend=False)
plt.title("Quantidade de Pedidos por Status")
plt.ylabel("Total de Pedidos")
plt.xlabel("Status do Pedido")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Merge com channels
pedidos_canais = dfs["orders"].merge(dfs["channels"], on="channel_id")

# Agrupando
canais = pedidos_canais["channel_name"].value_counts().head(10).reset_index()
canais.columns = ['channel_name', 'count']

# Gr√°fico
sns.barplot(data=canais, x='count', y='channel_name', palette="cubehelix", hue='channel_name', legend=False)
plt.title("Top 10 Canais de Venda por Quantidade de Pedidos")
plt.xlabel("Total de Pedidos")
plt.ylabel("Canal")
plt.tight_layout()
plt.show()

```

```python
# Faturamento total
faturamento_total = dfs["payments"]["payment_amount"].sum()
print(f"üí∞ Faturamento Total: R$ {faturamento_total:,.2f}")

# Faturamento por m√©todo de pagamento
faturamento_por_metodo = dfs["payments"].groupby("payment_method")["payment_amount"].sum().sort_values(ascending=False)

# Gr√°fico
faturamento_por_metodo.plot(kind="bar", color="steelblue")
plt.title("Faturamento por M√©todo de Pagamento")
plt.ylabel("Faturamento (R$)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Tempo total do ciclo de entrega (em minutos)
tempo_medio = dfs["orders"]["order_metric_cycle_time"].mean()
print(f"‚è±Ô∏è Tempo m√©dio de ciclo de pedido: {tempo_medio:.2f} minutos")
```

```python
# Merge orders -> stores -> hubs para ter cidade no pedido
orders_stores = dfs["orders"].merge(dfs["stores"][["store_id", "hub_id"]], on="store_id", how="left")
orders_hubs = orders_stores.merge(dfs["hubs"][["hub_id", "hub_city"]], on="hub_id", how="left")

# Contagem de pedidos por cidade
pedidos_por_cidade = orders_hubs["hub_city"].value_counts().reset_index()
pedidos_por_cidade.columns = ['hub_city', 'count']

# Gr√°fico
sns.barplot(data=pedidos_por_cidade, x='count', y='hub_city', palette="magma", hue='hub_city', legend=False)
plt.title("Quantidade de Pedidos por Cidade")
plt.xlabel("Total de Pedidos")
plt.ylabel("Cidade")
plt.tight_layout()
plt.show()

```

```python
# Pedidos por hora
pedidos_por_hora = dfs["orders"]["order_created_hour"].value_counts().sort_index()

sns.lineplot(x=pedidos_por_hora.index, y=pedidos_por_hora.values, marker="o")
plt.title("Pedidos por Hora do Dia")
plt.xlabel("Hora do Dia")
plt.ylabel("Total de Pedidos")
plt.grid(True)
plt.tight_layout()
plt.show()

# Pedidos por m√™s
pedidos_por_mes = dfs["orders"]["order_created_month"].value_counts().sort_index().reset_index()
pedidos_por_mes.columns = ['month', 'count']

sns.barplot(data=pedidos_por_mes, x='month', y='count', palette="coolwarm", hue='month', legend=False)
plt.title("Pedidos por M√™s")
plt.xlabel("M√™s")
plt.ylabel("Total de Pedidos")
plt.tight_layout()
plt.show()

```

```python
# Filtra pedidos cancelados
cancelados = dfs["orders"][dfs["orders"]["order_status"] == "CANCELED"]

# Merge com canais
cancelados_canais = cancelados.merge(dfs["channels"], on="channel_id")

# Contagem cancelamentos por canal
cancelamentos_canais = cancelados_canais["channel_name"].value_counts().reset_index()
cancelamentos_canais.columns = ['channel_name', 'count']

# Gr√°fico
sns.barplot(data=cancelamentos_canais, x='count', y='channel_name', palette="rocket", hue='channel_name', legend=False)
plt.title("Cancelamentos por Canal de Venda")
plt.xlabel("Total de Cancelamentos")
plt.ylabel("Canal")
plt.tight_layout()
plt.show()

```

```python
# Merge orders e stores
orders_stores_fat = dfs["orders"].merge(dfs["stores"][["store_id", "store_name"]], on="store_id")

# Agrupa por loja somando o valor do pedido (order_amount)
faturamento_lojas = orders_stores_fat.groupby("store_name")["order_amount"].sum().sort_values(ascending=False).head(10).reset_index()

# Gr√°fico
sns.barplot(data=faturamento_lojas, x="order_amount", y="store_name", palette="Blues_d", hue="store_name", legend=False)
plt.title("Top 10 Lojas por Faturamento")
plt.xlabel("Faturamento (R$)")
plt.ylabel("Loja")
plt.tight_layout()
plt.show()

```

```python
# Merge deliveries com drivers para juntar info do tipo de driver
deliveries_drivers = dfs["deliveries"].merge(dfs["drivers"], on="driver_id", how="left")

# Contagem de entregas por tipo de driver
entregas_por_tipo = deliveries_drivers["driver_type"].value_counts()

# Gr√°fico
sns.barplot(x=entregas_por_tipo.values, y=entregas_por_tipo.index, hue=entregas_por_tipo.index, palette="viridis", legend=False)
plt.title("Quantidade de Entregas por Tipo de Driver")
plt.xlabel("Total de Entregas")
plt.ylabel("Tipo de Driver")
plt.tight_layout()
plt.show()

```

```python
# M√©dia da dist√¢ncia percorrida por tipo de driver
distancia_media_tipo = deliveries_drivers.groupby("driver_type")["delivery_distance_meters"].mean().sort_values(ascending=False)

# Gr√°fico
sns.barplot(x=distancia_media_tipo.values, y=distancia_media_tipo.index, hue=distancia_media_tipo.index, palette="coolwarm", legend=False)
plt.title("Dist√¢ncia M√©dia de Entregas por Tipo de Driver")
plt.xlabel("Dist√¢ncia M√©dia (metros)")
plt.ylabel("Tipo de Driver")
plt.tight_layout()
plt.show()
```

```python
# Pivot table com contagem de status por tipo de driver
status_por_tipo = deliveries_drivers.pivot_table(index="driver_type", columns="delivery_status", aggfunc="size", fill_value=0)

# Visualiza√ß√£o com heatmap
sns.heatmap(status_por_tipo, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Status das Entregas por Tipo de Driver")
plt.xlabel("Status da Entrega")
plt.ylabel("Tipo de Driver")
plt.tight_layout()
plt.show()
```

```python
# Merge orders com stores para pegar o segmento
orders_stores_segment = dfs["orders"].merge(dfs["stores"][["store_id", "store_segment"]], on="store_id", how="left")

# Faturamento por segmento (soma do order_amount)
faturamento_segmento = orders_stores_segment.groupby("store_segment")["order_amount"].sum().sort_values(ascending=False)

# Quantidade de pedidos por segmento
pedidos_segmento = orders_stores_segment["store_segment"].value_counts()

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,5))

# Gr√°fico 1: Faturamento por segmento
plt.subplot(1, 2, 1)
sns.barplot(x=faturamento_segmento.values, y=faturamento_segmento.index, hue=faturamento_segmento.index, palette="coolwarm", dodge=False, legend=False)
plt.title("Faturamento por Segmento da Loja")
plt.xlabel("Faturamento (R$)")
plt.ylabel("Segmento")

# Gr√°fico 2: Quantidade de pedidos por segmento
plt.subplot(1, 2, 2)
sns.barplot(x=pedidos_segmento.values, y=pedidos_segmento.index, hue=pedidos_segmento.index, palette="viridis", dodge=False, legend=False)
plt.title("Quantidade de Pedidos por Segmento da Loja")
plt.xlabel("Total de Pedidos")
plt.ylabel("Segmento")

plt.tight_layout()
plt.show()

```
